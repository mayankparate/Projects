<font color="purple"><h4><center> **Performing Linear Regression Algorithm on our Dataset**</h4>

<font color="purple"><h4> **We are going to perform Linear Regression Algorithm on the column named GAD_T data which is the overall score of the General Anxiety Disorder. We are going to check how many players has anxiety by playing games on regular basis.**</h4>

<font color="purple"><h4> **We are going to perform linear regression to check the relation between game addiction and anxiety.**</h4>


<font color="purple">**We are going to perform algorithm on GAD_T column which is a total 
score of General Anxiety Disorder and Hours column.**

```python
#Grouping GAD_T and Hours column
add=data.groupby('GAD_T')['Hours'].mean()
add
```

<font color="purple">**Making CSV file.**


```python
add.to_csv('add')
```

<font color="purple">**Reading csv file**

```python
add=pd.read_csv("/content/add")
add
```

<font color="purple">**Plotting bar graph of the above csv file.**

```python
add.plot(kind="scatter",y="GAD_T",x="Hours")
plt.show()
```

```python
sns.kdeplot(data=add,shade=True)
plt.show()
```

<font color="purple">**Now starting the prediction by using Linear Regression Algorithm.**

<font color="purple">**Storing the independent variable and dependent variable into x and y respectively.**

```python
x=add.iloc[:,0].values   #store independent variable
y=add.iloc[:,1].values   #store dependent variable
```

```python
print("Value of x is :- ",x)
print("Value of y is :- ",y)
```

<font color="purple">**Reshaping the values of x and y**

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
x
```

```python
y
```

<font color="purple">**Splitting the data into train and test data set.**


```python
from sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20) 
#it means data divide into 80/20 %
```

```python
print("Size of xtrain :- ",xtrain.size)
print("Size of xtest :- ",xtest.size)
print("Size of ytrain :- ",ytrain.size)
print("Size of ytest :- ",ytest.size)
```

```python
print("value of ytest :-",ytest)
print("value of xtest :-",xtest)
```

<font color="purple">**Preparing Linear Regression Model and Fitting it**

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
print("Intercept :-",model.intercept_)
print("slope :- ",model.coef_)
```

<font color="purple">**Predicting response**

```python
ypred = model.predict(xtest)
ypred
```

<font color="purple">**Getting Result**

<font color="purple">**Calculating the R-square value of ytest and ypred**

```python
#from typing import ValuesView
#Calculate R-Square ValuesView 
from sklearn.metrics import r2_score
r2=r2_score(ytest,ypred)
print(f"coefficient of determination: {r2}")
```

<font color="purple"> **Calculating Mean Sqaured error**

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(ytest,ypred)
```

```python
print('Variance score: %.2f' % model.score(xtest,ytest))
```

```python
#y=mx+c
m=model.coef_
c=model.intercept_
print("m=",m)
print("c=",c)
```

<font color="purple">**Using Linear regression formula**

```python
y1=m*xtest+c
```

```python
y1
```

<font color="purple">**Plotting scatter graph of training models and plotting regression line of xtest model and its predicted value**

```python
plt.style.use('fivethirtyeight')
plt.scatter(xtrain,ytrain,color="green")
plt.plot(xtest,model.predict(xtest),color="blue")
plt.xlabel("GAD_T Score")
plt.ylabel("Hours")
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.scatter(model.predict(xtrain), model.predict(xtrain) - ytrain,    ## plotting residual errors in training data
            color = "green", s = 10, label = 'Train data')  
plt.scatter(model.predict(xtest), model.predict(xtest) - ytest,       ## plotting residual errors in test data
            color = "blue", s = 10, label = 'Test data')
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
plt.legend(loc = 'upper right')
plt.title("Residual errors")
plt.show()
```
